```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('train.csv')
df
```

```python
df.tail(20)
```

```python
df['Title'] =  df['Name'].str.extract('([A-Za-z]+)\.')  ## ([A-Za-z]+)\. this is just normal english jese A kaa matlab hai first letter capital then Za ka matlab capital and small ka combo and z ka matlab hai small at the end
```

```python
df['Title'].value_counts()
```

```python
def replace_title(row):
    title = row["Title"]
    relevant_title = ['Mr','Mrs','Miss','Master']
    if title in relevant_title:
        return title
    else:
        return "Others"
    
df['Title'] = df.apply(replace_title,axis=1)

df
```

```python
df['Title'].unique()
```

```python
df[df["Title"]=="Mr"]["Age"].mean()
```

```python
palette = ['red', 'green', 'blue', 'orange', 'purple']
plt.figure(figsize=(8, 5))
sns.countplot(data=df[df['Survived'] == 1], x='Title', palette=palette)
plt.title('Survivors by Title')
plt.xlabel('Title')
plt.ylabel('Number of Survivors')
plt.show()
```

```python
sns.barplot(data = df, x = 'Title',y='Survived')
plt.title('Survivors by Title')
plt.show()
```
